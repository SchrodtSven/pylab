## [PEP 655](https://peps.python.org/pep-0655/): Marking individual TypedDict items as required or not-requiredÂ¶

```python
from typing import TypedDict, Required, NotRequired


class Movie(TypedDict):
   title: str
   year: NotRequired[int]

m1: Movie = {"title": "Amazing Fantasy #15", "year": 1962}  # OK
m2: Movie = {"title": "Captain Marvel"}  # OK (year is not required)
#m3: Movie = {"year": 2022}  # ERROR (missing required field title)


print([m1, m2])
```
