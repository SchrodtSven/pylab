# Old(er) string operating stuff

## The String Formatting Operator (%)

Inspired by *printf* function in C-like languages

```python
city = "Moers"
temperature = -2
template = "The temperature in %s is %d °C. That's %s"

template % (city, temperature, 'cold')
```

## The `str.format()` Method

```python
tpl = "Hello {0}, let's talk about {1} - {0}?"
tpl.format('alien life form', 'catz')
```

## The `string.Template` Class

```python
from string import Template

template = Template("Hello, $name! Today is $day.")
template.substitute(name="Pythonista", day="Friday")
```

The 'string.Formatter.parse()` Method

```python
from string import Formatter

template = "The temperature in {city!r} is {temperature:.2f} F."

for text, field, format_spec, conversion in Formatter().parse(template):
    print(f"{text=}, {field=}, {format_spec=}, {conversion=}")
```

## F-Strings

In f-strings, Python supports three conversion flags, `!s`, `!r`, and `!a`<sup>*</sup>, that let you decide which string representation to use when interpolating objects. If you specify a conversion flag in an f-string, the result of evaluating the expression is converted to the specified string representation.

--- 
<sup>*</sup>
`!s` which calls `str()` on the value, `!r` which calls `repr()` and `!a` which calls `ascii()`
